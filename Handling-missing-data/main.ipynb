```python
import pandas as pd

try:
  # Read the CSV file
  df = pd.read_csv('blood_data_set.csv')

  # Select only the desired columns: 'Creatinine' and 'Glucose'
  clean_df = df[['Creatinine', 'Glucose']]

  # Filter for rows where 'Disease' is 'Diabetes'
  clean_df = clean_df[clean_df['Disease'] == 'Diabetes']

  # Print or save the cleaned DataFrame
  print(clean_df)

  # To save the cleaned data to a new CSV file:
  clean_df.to_csv('diabetes_data.csv', index=False)  # Replace 'diabetes_data.csv' with your desired filename

except FileNotFoundError:
  print("Error: 'blood_data_set.csv' file not found.")

except KeyError:
  print("Error: 'Disease' or 'Creatinine' or 'Glucose' column not found in the DataFrame.")
```
